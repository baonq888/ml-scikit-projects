### 2.1 Split Train/Test Data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    cancer['data'], cancer['target'], random_state=5)

X_train_log = np.log1p(X_train)
X_test_log = np.log1p(X_test)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_log)
X_test_scaled = scaler.transform(X_test_log)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

### 2.2 Overview

```python
import pandas as pd
import mglearn # type: ignore
# create dataframe from data in X_train
# label the columns using the strings in iris_dataset.feature_names
cancer_dataframe = pd.DataFrame(X_train, columns=cancer.feature_names)
display(cancer_dataframe[:10])
```

Explanation on each tumour measurement defintions


| **Feature**             | **Explanation**                                                                 |
|--------------------------|----------------------------------------------------------------------------------|
| **Radius**              | Distance from the center to the edge of the nucleus. Represents cell size.       |
| **Texture**             | Variation in pixel intensity. Describes surface roughness or internal patterns.  |
| **Perimeter**           | Length around the boundary of the nucleus.                                       |
| **Area**                | Total area inside the nucleus (in pixels).                                       |
| **Smoothness**          | Measures how smooth the boundary is. Lower = smoother edge.                      |
| **Compactness**         | Indicates how compact vs. spread out the nucleus shape is.                       |
| **Concavity**           | How deep the concave (inward) curves are along the nucleus boundary.             |
| **Concave Points**      | Number of concave sections (indentations) on the nucleus edge.                   |
| **Symmetry**            | Degree of symmetry of the nucleus. Benign tumors tend to be more symmetrical.    |
| **Fractal Dimension**   | Measures the complexity of the boundary (higher = more irregular).               |

### 2.3 Benign vs Malignant

```python
y = cancer['target']

target_counts = pd.Series(y).value_counts()

percentage = target_counts / len(y) * 100

result = pd.DataFrame({
    'Diagnosis': ['Benign (Yes)', 'Malignant (No)'],
    'Count': target_counts.values,
    'Percentage': percentage.values
})

result
```

It is clear that the percentage of Benign tumour accounts for about 62% while Malignant tumour accounts for about 37%, with a gap of over 100 rows, making the data slightly skewed.

It is probable that mean symmetry is less helpful in distinguishing tumor types, while others may contribute largely to classify tumour types.

### Quantify Outliers

```python
Q1 = cancer_dataframe.quantile(0.25)
Q3 = cancer_dataframe.quantile(0.75)
IQR = Q3 - Q1

outliers = ((cancer_dataframe < (Q1 - 1.5 * IQR)) | (cancer_dataframe > (Q3 + 1.5 * IQR)))
outlier_counts = outliers.sum().sort_values(ascending=False)
print(outlier_counts)
```

### Measure Feature Importance

Point Biserial Correlation

```python
from scipy.stats import pointbiserialr

correlations = {}
for col in cancer_dataframe.columns:
    corr, _ = pointbiserialr(cancer_dataframe[col], y_train)
    correlations[col] = corr

corr_df = pd.DataFrame(correlations.items(), columns=["Feature", "PointBiserialCorr"])
corr_df = corr_df.sort_values(by="PointBiserialCorr", key=abs, ascending=False)

corr_df.head(10)
```

It is shown that these features have more malignant tumours than bengin

### Top Features to Explore

```python
top_10_features = corr_df.head(10)['Feature'].tolist()
top_10_features
```

```python
cancer_df = pd.DataFrame(X_train, columns=cancer.feature_names)
cancer_df['target'] = y_train

# Group by class and describe top features
class_stats = cancer_df.groupby('target')[top_10_features].describe().transpose()
print(class_stats)
```

## 3. Training

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg100 = LogisticRegression(C=100, max_iter=100).fit(X_train_scaled, y_train)

print("Training set score: {:.3f}".format(logreg100.score(X_train_scaled, y_train)))
print("Test set score: {:.3f}".format(logreg100.score(X_test_scaled, y_test)))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train_scaled, y_train)
print("Accuracy on training set: {:.3f}".format(tree.score(X_train_scaled, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test_scaled, y_test)))
```

```python
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    
plot_feature_importances_cancer(tree)
```
